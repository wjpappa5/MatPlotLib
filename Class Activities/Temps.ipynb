```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(1,13,1)
y_axis = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
celcius = [(y - 32) * 0.56 for y in y_axis]
celcius
```

```python
plt.plot(x_axis, celcius)
plt.show()
```

```python
plt.plot(x_axis, y_axis)
plt.plot(x_axis, celcius)
plt.show()
```

```python
F_handle, = plt.plot(x_axis, y_axis, marker ='+', color='blue', label="Fahreneit")
C_handle, = plt.plot(x_axis, celcius, marker='o', color='red', label="Celcius")
plt.legend(loc="upper right")
plt.title("New Jersey Temperatures")
plt.xlabel("Month")
plt.ylabel("Temperature")
```
