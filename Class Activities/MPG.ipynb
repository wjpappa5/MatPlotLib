```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "/Users/williampappas/Desktop/GitLab/01-Class-Activities/05-Matplotlib/3/Activities/05-Stu_MilesPerGallon-ScatterPlot/Resources/mpg.csv"
mpg_df = pd.read_csv(file)
mpg_df.head()
```

```python
mpg_df = mpg_df.loc[mpg_df['horsepower'] != "?"]
mpg_df.head()
```

```python
mpg_df = mpg_df.set_index("car name")
del mpg_df['origin']
mpg_df.head()
```

```python
mpg_df['horsepower']=pd.to_numeric(mpg_df["horsepower"])
```

```python
mpg_df.plot(kind="scatter",x="horsepower", y="mpg", grid=True,title="Horsepower vs. MPG")
```
