```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_df = pd.read_csv("/Users/williampappas/Desktop/GitLab/01-Class-Activities/05-Matplotlib/2/Activities/08-Stu_BattlingKings-PlottingPandas/Unsolved/Resources/got.csv")
got_df.head()
```

```python
# Get attacker and defender data
attackers=got_df["attacker_king"].value_counts()
defenders=got_df["defender_king"].value_counts()
```

```python
# Get total battle data
battle_data = attackers.add(defenders, fill_value=0)
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar",facecolor="red")
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Fought")
plt.xlabel("King")
```
